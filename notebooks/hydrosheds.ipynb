```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

rgi_shapefile_path = "/home/bsurya/Projects/AIR-Zones/data/leh_glims_download_67564/glims_download_67564/glims_polygons.shp"

rgi_glaciers = gpd.read_file(rgi_shapefile_path)

# Bounding box coordinates for Ladakh region
xmin = 77.5
ymin = 35
xmax = 78.5
ymax = 34

# Read the shapefile into a geopandas dataframe
ladakh_rivers = rivers[rivers.ORD_STRA==1]

# Create the plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# Set the extent of the plot
ax.set_extent([xmin, xmax, ymin, ymax], crs=ccrs.PlateCarree())

# Add Natural Earth features as the basemap
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='white')
ax.add_feature(cfeature.COASTLINE, linewidth=0.5, edgecolor='black')
ax.add_feature(cfeature.BORDERS, linewidth=0.5, edgecolor='black')

# Plot the rivers in blue
ladakh_rivers.plot(ax=ax, color='blue', linewidth=1, transform=ccrs.PlateCarree())

# Plot the glacier outlines from RGI
rgi_glaciers.plot(ax=ax, facecolor='none', edgecolor='cyan', linewidth=0.5, transform=ccrs.PlateCarree())

# Indicate the capital city of Leh
leh_lon = 77.577049  # Longitude of Leh
leh_lat = 34.164840  # Latitude of Leh
ax.plot(leh_lon, leh_lat, 'ro', markersize=8, transform=ccrs.PlateCarree())  # Plotting Leh as a red dot

# Add labels, title, etc. if desired
ax.set_title('Ladakh Region with Rivers, Glacier Outlines, and Leh')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Display the plot
plt.show()
```

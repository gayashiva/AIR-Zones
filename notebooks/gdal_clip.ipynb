```python
from osgeo import gdal
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ds = gdal.Open("../data/HydroRIVERS_v10_shp/raster/reduced_HydroRIVERS_v10.tif")
# visualize
array = ds.GetRasterBand(1).ReadAsArray()
plt.figure()
plt.imshow(array)
plt.colorbar()
plt.show()
```

```python
# make sure your raster data and shapefile have the same projection!
dsClip = gdal.Warp("../data/demClip.tif", ds, cutlineDSName = "../data/K3Binary/k3binary.shp",
                   cropToCutline = True, dstNodata = np.nan)
```

```python
# visualize
array = dsClip.GetRasterBand(1).ReadAsArray()
plt.figure()
plt.imshow(array)
plt.colorbar()
plt.show()
```

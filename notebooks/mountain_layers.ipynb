```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Read the RiverAtlas Shapefile
mountainfile_path = 'data/K3Binary/k3binary.shp'

df = gpd.read_file(mountainfile_path)
df.head()
```

```python
# Read the RiverAtlas Shapefile
dischargefile_path = 'data/HydroRIVERS_v10_shp/clip_HydroRIVERS_v10/clip_HydroRIVERS_v10.shp'

dis_df = gpd.read_file(dischargefile_path)
dis_df.head()
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Create a figure and subplot
fig, ax = plt.subplots(figsize=(12, 6), subplot_kw={'projection': ccrs.PlateCarree()})

# Add land and ocean background
ax.add_feature(cfeature.LAND, facecolor='none')

# Plot the RiverAtlas data
df.plot(color='lightgray',ax=ax)

# Plot the RiverAtlas data
dis_df.plot(color='red',ax=ax)

# Display the plot
plt.show()
```
